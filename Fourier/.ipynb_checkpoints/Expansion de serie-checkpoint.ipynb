<h3>Expansión de serie de Fourier</h3>

```python
from IPython.display import HTML, Math, Latex, display
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
from sympy.printing import latex
import math

n, t = symbols('n,t')

T = 2*pi
w = 1
a0 = (4*pi**2)/3
an = (4/n**2)
bn = (-4*pi/n)
expresion = an * cos(n*w*t) + bn * sin(n*w*t)
expansion = a0

writeable_strs = ["$$g(t) = {}$$".format(latex(a0))]
for i in range(1, 20):
    next = expresion.subs(n, i)
    expansion = next + expansion
    writeable_strs.append("$${}$$".format(latex(next)))
```

Para copiar:

```python
for i in writeable_strs:
    display(Latex(i))
```

Para desmos:

```python
def desmos_syntax(e):
    return str(e).replace("**", "^").replace("pi", "π")
desmos_str = "g=" + desmos_syntax(expansion)
print(desmos_str)
```

Latex continuo:

```python
latex_str = "$$g(t) = {}$$".format(latex(expansion))
display(Latex(latex_str))
```
