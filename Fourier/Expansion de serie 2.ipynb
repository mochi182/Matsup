<h3>Expansión de serie de Fourier</h3>

```python
from IPython.display import HTML, Math, Latex, display
from sympy import *
from sympy.printing import latex

n, t = symbols('n,t')

T = 2*pi
w = 1
a0 = pi/2
an = -2/(pi*(n**2))
bn = -((-1)**n)/n
expresion_1 = an * cos(n*w*t)
expresion_2 = bn * sin(n*w*t)
expansion = a0/2

writeable_strs = ["$$g(t) = {}$$".format(latex(expansion))]
for i in range(10):
    next = expresion_1.subs(n, (2*n-1)).subs(n, i+1) + expresion_2.subs(n, i+1)
    expansion = next + expansion
    writeable_strs.append("$${}$$".format(latex(next)))

```

Para copiar:

```python
for i in writeable_strs:
    display(Latex(i))
```

Para desmos:

```python
def desmos_syntax(e):
    return str(e).replace("**", "^").replace("pi", "π")
desmos_str = "g=" + desmos_syntax(expansion)
print(desmos_str)
```

Latex continuo:

```python
latex_str = "$$g(t) = {}$$".format(latex(expansion))
display(Latex(latex_str))
```
